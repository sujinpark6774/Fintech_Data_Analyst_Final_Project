```python
from konlpy.tag import Twitter
from collections import Counter
import os
import pandas as pd


# 파일 불러오기
os.chdir(r'C:\Users\LGR7\Desktop\핀테크')

hana = pd.read_csv('hn_total.csv',encoding = 'cp949')
hana_title = hana['text']

# 이벤트 제목 리스트화
list_data=list()
for i in range(len(hana_title)):
    list_data.append(hana_title.loc[i])

# 형태소 분석
twitter = Twitter() 

# 형태소 분석 결과
morphs = [] 

for sentence in list_data: 
    morphs.append(twitter.pos(sentence))     
#print(morphs)


noun_adj_adv_list=[]

for sentence in morphs : 
    for word, tag in sentence : 
        if tag in ['Noun'] and ("것" not in word)and("이벤트"not in word)and("혜택"not in word)and("최대"not in word)and("프로모션"not in word)and("할인"not in word) and ("내" not in word)and ("나" not in word)and ("수"not in word) and("게"not in word)and("말"not in word): 
            noun_adj_adv_list.append(word)             
#print(noun_adj_adv_list)

# 단어 빈도
count = Counter(noun_adj_adv_list)
words = dict(count.most_common())
#print(words)

# TOP 10
import operator

top10=sorted(words.items(),key=operator.itemgetter(1), reverse=True)
top10[:100]
#print(top10)

# 시각화

from wordcloud import WordCloud 
import matplotlib.pyplot as plt 
import nltk 
from nltk.corpus import stopwords
%matplotlib qt
# %matplotlib inline
import matplotlib 
from IPython.display import set_matplotlib_formats 


wordcloud = WordCloud(font_path = 'C:/Windows/Fonts/HMKMG.ttf', background_color='white',colormap = "Accent_r", width=1500, height=1000).generate_from_frequencies(words) 
plt.imshow(wordcloud) 
plt.axis('off') 
plt.show()
```
