```python
## konelpy 패키지중 하나
from konlpy.tag import Twitter
## counter 단어 카운팅 패키지 
from collections import Counter
import os
import pandas as pd
import re

from wordcloud import WordCloud

import matplotlib.pyplot as plt
from PIL import Image                         # Pillow 패키지의 영상 핸들링 클래스.
import matplotlib 
from IPython.display import set_matplotlib_formats
#from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
%matplotlib inline

twitter = Twitter()


os.chdir(r'C:\Users\LGR7\Desktop\핀테크')

# 1)파일 불러오기
test=pd.read_csv("hn_total.csv",encoding="cp949")

# 2) 전처리
#### 리스트로 변경
re_data=[]

for a_line in test['text']:
    pre = re.sub(r'\W', ' ', a_line)                                # 특수문자 제외.
    pre = re.sub(r'_', ' ', pre)                                    # 특수문자 제외.
    #pre = re.sub(r'\d+','', pre)                                    # 수자 제외.
    re_data.append(pre)


# 3) 형태소 분석
morphs = [] 

for sentence in re_data: 
    morphs.append(twitter.pos(sentence))     
#print(morphs)

### tag 명사 추출, word에서 단글자 제거(수동)
noun_adj_adv_list=[]
for sentence in morphs : 
    for word, tag in sentence : 
        if tag in ['Noun'] and ("것" not in word) and ("와" not in word)and ("나" not in word)and ("더"not in word) and("고"not in word)and("해"not in word): 
            noun_adj_adv_list.append(word)             
#print(noun_adj_adv_list)

### 입력 단어 제거 ###################
final_words=[]

### if (제거단어 not in word) and (제거단어 not in word) and (제거단어 not in word) :
for word in noun_adj_adv_list :
    if ('이벤트' not in word):
        final_words.append(word)

# 4)단어 빈도
count = Counter(final_words)
words = dict(count.most_common())
#print(words)

## TOP 10
import operator

top10=sorted(words.items(),key=operator.itemgetter(1), reverse=True)
## 숫자 확인하고 싶은 것으로 수정가능
top10[:50]


# 5) 시각화

from wordcloud import WordCloud 
import matplotlib.pyplot as plt 
import nltk 
from nltk.corpus import stopwords
%matplotlib inline 
import matplotlib 
from IPython.display import set_matplotlib_formats 

## 기본셋팅
# 폰트
matplotlib.rc('font',family = 'Malgun Gothic') 
# 폰트 선명하게
set_matplotlib_formats('retina') 
# 그래프 음수 수치 오류 방지
matplotlib.rc('axes',unicode_minus = False)

## 실행
wordcloud = WordCloud(font_path = 'C:/Windows/Fonts/malgun.ttf', background_color='white',colormap = "Accent_r", width=1500, height=1000).generate_from_frequencies(words) 
plt.imshow(wordcloud) 
plt.axis('on') 
plt.show()


######## 검색 키워드 제거해야함 ######################## ex) 여행
#########
```
